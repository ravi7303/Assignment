**4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.**

**Ans-**we can use a box plot (also known as a box-and-whisker plot). A box plot shows the distribution of a numerical variable within each category, including the median, quartiles, and potential outliers.

Steps:

1.Generate a dataset with categories and corresponding numerical values.

2.Visualize the distribution of the numerical variable across different categories using a box plot.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic dataset
np.random.seed(42)

# Categories
categories = ['Category A', 'Category B', 'Category C', 'Category D']

# Generate numerical values for each category (random data for example)
data = {
    'Category A': np.random.normal(loc=50, scale=10, size=100),
    'Category B': np.random.normal(loc=60, scale=15, size=100),
    'Category C': np.random.normal(loc=70, scale=20, size=100),
    'Category D': np.random.normal(loc=65, scale=5, size=100)
}

# Combine the data into a list of arrays (one for each category)
values = [data[category] for category in categories]

# Create box plot
plt.boxplot(values, vert=True, patch_artist=True,
            labels=categories,
            boxprops=dict(facecolor='skyblue', color='black'),
            medianprops=dict(color='red'))

# Add titles and labels
plt.title("Distribution of Numerical Values Across Categories")
plt.xlabel("Categories")
plt.ylabel("Numerical Value")

# Show plot
plt.show()
```

**5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap**.

**Ans-** To generate a synthetic dataset with correlated features and visualize the correlation matrix using a heatmap, we can follow these

steps:

1.Generate a synthetic dataset with correlated features.

2.Compute the correlation matrix between the features.

3.Visualize the correlation matrix using a heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate synthetic dataset with correlated features
np.random.seed(42)

# Create random data with correlation
n_samples = 100
x = np.random.rand(n_samples)  # Feature 1
y = 2 * x + np.random.randn(n_samples) * 0.5  # Feature 2, correlated with x
z = -x + np.random.randn(n_samples) * 0.5  # Feature 3, negatively correlated with x

# Combine into a DataFrame
data = pd.DataFrame({'X': x, 'Y': y, 'Z': z})

# Step 2: Compute the correlation matrix
correlation_matrix = data.corr()

# Step 3: Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt='.2f', cbar=True)

# Add title
plt.title("Correlation Matrix Heatmap")

# Show plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Generate the dataset
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Step 2: Create the bubble chart
plt.figure(figsize=(10, 6))

# Scatter plot: x = GDP, y = Population, size of bubbles = Population
plt.scatter(df['GDP'], df['Population'], s=df['Population']*3, alpha=0.6, c='blue', edgecolors="w", linewidth=2)

# Step 3: Add labels and customize the plot
for i, country in enumerate(df['Country']):
    plt.text(df['GDP'][i] + 20, df['Population'][i], country, fontsize=12, ha='center')

# Add titles and labels
plt.title("Bubble Chart of Population vs GDP by Country", fontsize=16)
plt.xlabel("GDP (in billions)", fontsize=14)
plt.ylabel("Population (in millions)", fontsize=14)

# Show grid
plt.grid(True)

# Show the plot
plt.show()
```
